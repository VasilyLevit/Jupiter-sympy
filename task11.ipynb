Дана функция f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *
```

```python
x = symbols('x')
```

```python
f = - 12*x**4*sin(cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30
interval = Interval(-5, 5)
```

```python
solve(5*x**2 + 10*x - 30)
```

```python
list(map(N, solveset(5*x**2 + 10*x - 30, x)))
```

```python
solveset(5*x**2 + 10*x - 30, x)
```

```python
from sympy.plotting import plot
plot(5*x**2 + 10*x - 30, x)
```

```python
minimum(5*x**2 + 10*x - 30, x)
```

```python
minimum(5*x**2 + 10*x - 30, x, S.Reals)
```

```python
solveset(f, x, interval)
```

```python
from sympy.plotting import plot
plot(f, (x, -30, 30))
```

```python
minimum(f, x, S.Reals)
```

```python
diff(5*(x**2) + 10*x - 30, x)
```

```python
diff(f, x)
```

```python
solveset(f.diff(x), x, interval)
```

```python
is_increasing(5*x**2 + 10*x - 30, Interval.open(-1, oo))
```

```python
is_decreasing(5*x**2 + 10*x - 30, Interval.open(-oo, -1))
```
